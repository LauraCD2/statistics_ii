#Distribución normal

```python
def fundDensidad(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

def dist_normal(x, mu, sigma):
    prob, _ = integrate.quad(fundDensidad, -np.inf, x, args=(mu, sigma)) #calcula la integral definida
    return prob
```

#Teorema del límite central
El teorema del límite central (TCL) establece que, dada una muestra aleatoria suficientemente grande de la población, la distribución de las medias muestrales seguirá una distribución normal, independientementede la forma de la distribución de la población.

```python
data = np.arange(1, 10001)
n = 1000
n_muestra = 50
medias = []

for i in range(n):
    random_muestra = np.random.choice(data, n_muestra)
    media = np.mean(random_muestra)
    medias.append(media)

plt.figure(figsize=(10, 5))
sns.histplot(medias, bins=30, kde=True, color='b')
plt.xlabel('medias')
plt.ylabel('Frecuencia')
plt.title('TLC')
plt.show()
```

```python
def prob_TLC(x, mu, sigma, n):
    sigma_x_barra = sigma / np.sqrt(n)

    prob = dist_normal(x, mu, sigma_x_barra)
    return prob
```

#Prueba de hipótesis
## Inferencia sobre la media de una población: varianza conocida

Usamos la distribución normal estándar (Z) para calcular el estadístico de prueba:

$$Z=\dfrac{\bar{X}-\mu}{\sigma/\sqrt{n}}$$



```python
def prueba_hipotesis_completa(x_barra, mu_0, sigma, n, alpha, cola):
    Z = (x_barra - mu_0) / (sigma / np.sqrt(n))


    if cola == 'inf':
      z_medio = norm.ppf(1 - alpha)
      if Z < z_medio:
        print('Se rechaza Ho bajo', alpha)
      else:
        print('No se puede rechazar Ho bajo', alpha)


    if cola == 'sup':
      z_medio = norm.ppf(alpha)
      if Z > z_medio:
        print('Se rechaza Ho bajo', alpha)
      else:
        print('No se puede rechazar Ho bajo', alpha)


    if cola == 'dos':
      z_medio = norm.ppf(1 - alpha / 2)
      if abs(Z) > z_medio:
        print('Se rechaza Ho bajo', alpha)
      else:
        print('No se puede rechazar Ho bajo', alpha)
```

#Ley de los grandes números
La Ley de los Grandes Números establece que, a medida que aumenta el número de observaciones en un experimento aleatorio, la media muestral tiende a acercarse a la media real (poblacional) de la distribución.

#Ejercicios:

7.- Un servidor de bases de datos recibe solicitudes de consulta de usuarios. El tiempo de respuesta de cada consulta varía debido a la carga del sistema y la optimización del código. Se ha observado que, en promedio, el tiempo de respuesta esperado es de 200 ms, pero hay fluctuaciones.

Un administrador monitorea el servidor y registra los tiempos de respuesta de las consultas. Si toma muestras de 100, 500 y 5000 tiempos de respuesta, ¿cómo se comporta la media muestral en comparación con los 200 ms esperados?

```python
#ejercicio 7

mu = 200
sigma = 50
muestras = np.arange(10, 10000, 10)
medias = []

for n in muestras:
    tiempos_respuesta = np.random.normal(mu, sigma, n)
    media_muestral = np.mean(tiempos_respuesta)
    medias.append(media_muestral)

plt.figure(figsize=(10, 6))
plt.plot(muestras, medias)
plt.axhline(mu, color='r', linestyle='dashed', linewidth=2)
plt.xlabel('n')
plt.ylabel('Media')
plt.title('LGN')
plt.show()
```

A medida que el número de muestras aumenta, el promedio de los tiempos registrados se acercará a 200, confirmando la ley de los grandes números.
